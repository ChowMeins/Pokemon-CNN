```python
import json

with open("data_generation/pokemon_names.json", "r") as f:
    pokemon_names = json.load(f)
```

```python
import requests
from pathlib import Path
from bs4 import BeautifulSoup

base_url = "http://pokemondb.net/sprites"
variations = {}
for i, name in enumerate(pokemon_names):
    request = requests.get(f"{base_url}/{name}")
    print(name)
    html = request.text
    soup = BeautifulSoup(html, "html.parser")
    gen4_heading = soup.find("h2", string="Generation 4")
    gen4_table = gen4_heading.find_next("tbody")
    gen4_sprite_columns = gen4_table.find_all("tr")
    for col in gen4_sprite_columns:
        src = col.find_all("img")
        for s in src:
            if i + 1 not in variations:
                variations[i + 1] = set()
            variations[i + 1].add(str(Path(s["src"]).stem))

```
